```python
import torch
from diffusers import DiffusionPipeline
from diffusers.utils import load_image

# switch to "mps" for apple devices
pipe = DiffusionPipeline.from_pretrained("Qwen/Qwen-Image-Edit-2509", torch_dtype=torch.bfloat16, device_map="cuda")

prompt = "Change the outfit to bikini"
input_image = load_image("https://www.kerala9.com/wp-content/gallery/keerthi-suresh/keerthy-suresh-in-black-off-shoulder-gown-photos-001.jpg")

image = pipe(image=input_image, prompt=prompt).images[0]
```
